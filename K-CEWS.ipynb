## **2.0 Introduction: The 14-Day Race**
> *"In public health, time is the only currency that matters. A 24-hour delay in response can result in an exponential surge in infections; a 14-day head start can save a city."*

Cholera remains a persistent shadow over Kenya’s public health landscape, particularly within the Lake Victoria Basin and urban informal settlements. Despite significant advances in medical treatment, the **National Multi-sectoral Cholera Elimination Plan (2022-2030)** identifies a critical failure: our surveillance systems are reactive. We detect outbreaks only after hospitals are overwhelmed.

**K-CEWS (Kenya Cholera Early Warning System)** flips this script. By utilizing **spatiotemporal machine learning**, we bridge the gap between "Climate Shocks" (heavy rainfall and flooding) and the biological transmission cycle of *Vibrio cholerae*. K-CEWS provides a predictive "Sentinel" that looks 14 days into the future, allowing the Ministry of Health to move from managing crises to preventing them.


## **3.0 Problem Statement**
The current epidemiological surveillance in Kenya relies on the Integrated Disease Surveillance and Response (IDSR) framework. While robust for tracking, it lacks predictive foresight.

* **The Lag Gap:** There is a documented correlation between hydro-climatic triggers (rainfall) and cholera surges, yet this data is not integrated into real-time health alerts.
* **Resource Inefficiency:** Emergency response (vaccine deployment and WASH kits) is currently triggered by "Case Spikes," which means resources often arrive after the peak of the transmission curve.
* **Data Silos:** Environmental data (NASA), health data (WHO), and demographic data (KNBS) exist in isolation, preventing a holistic understanding of outbreak risk.


## **4.0 Aims and Objectives**

## **7.0 Scope and Limitations**
* **Scope:** The project focuses on high-risk reservoirs including the Lake Victoria Basin (Kisumu, Migori, Homa Bay) and Nairobi’s informal settlements. Data spans 2018 to April 2026.
* **Limitations:** Accuracy is dependent on the consistency of reporting in the KHIS (DHIS2) system. Unexpected human migrations (refugee movements) may introduce variables outside of climatic triggers.


## **8.0 Definition of Key Terms**
* **Spatiotemporal:** Data that encompasses both geographic location (Space) and time-stamped intervals (Time).
* **Feature Lagging (t-14):** The process of shifting independent variables (rainfall) back by 14 days to align with the future dependent variable (cholera cases).
* **Overdispersion:** A statistical property where the variance is higher than the mean, common in epidemic data where cases "spike" suddenly.


## **9.0 Ethical Considerations & Data Governance**
The development of K-CEWS adheres to the **Data Protection Act (2019) of Kenya** and international ethical standards.

### **9.1 De-identification and Anonymization**
* **The K-CEWS Protocol:** We operate strictly at the Sub-County/Aggregate level. The model does not ingest or store "Line Lists" containing names or exact household GPS coordinates. By using aggregated incidence rates (cases per 100k), we protect individual privacy.

### **9.2 Avoiding Community Stigmatization**
* **The K-CEWS Protocol:** Our dashboard is designed as a Decision-Support Tool for Health Officials, not a public-facing map. Access is restricted to authorized SCHMT members to ensure data is used for intervention rather than isolation.

### **9.3 Algorithmic Bias & Equity**
* **The K-CEWS Protocol:** We utilize Negative Binomial Regression to handle sparse data and incorporate KNBS socio-demographic baselines to ensure low-reporting rural areas are not ignored by the model.

## **Chapter 2: Literature Review**

### **2.1 Global Benchmark: The NASA-UNICEF Yemen Project**
* **The Project:** A high-level collaboration using satellite data to predict cholera in Yemen.
* **Successes:** Achieved 92% accuracy using NASA environmental "signatures."
* **Gaps:** Faced a "War-Data Gap"—unreliable ground reporting prevented real-time verification.
* **K-CEWS Capitalization:** Kenya’s stable KHIS/DHIS2 reporting allows for superior model verification compared to the Yemen study.

### **2.2 Local Benchmark: The Kenya Red Cross sEAP Protocol**
* **The Project:** The Simplified Early Action Protocol (sEAP) for Cholera in Kenya.
* **Successes:** Established clear environmental triggers (e.g., rainfall >50mm/day).
* **Gaps:** Relies on manual checks of rain gauges, making it "Human-Heavy" and slow.
* **K-CEWS Capitalization:** We are automating these triggers using Machine Learning for instantaneous, data-driven alerts.

### **2.3 Technical Benchmark: Nigeria Machine Learning Study (2026)**
* **The Project:** "Optimizing Cholera Outbreak Prediction in Nigeria."
* **Successes:** Proved XGBoost explained 69% of case variance.
* **Gaps:** Lacked "Lead Time"—models predicted the *current* week rather than providing a future window.
* **K-CEWS Capitalization:** By adding the **14-day Feature Lag**, we provide the "Foresight Window" that previous studies lacked.

### **2.4 Summary of Gaps & K-CEWS Innovation**

| Limitation in Previous Work | K-CEWS Solution (Our Capitalization) |
| :--- | :--- |
| **Reactive Monitoring:** Models only show current status. | **Predictive Lag:** Implementing a t-14 day lag for early warning. |
| **Environmental Isolation:** Models ignore population density. | **Holistic Integration:** Combining NASA (Rain) + KNBS (Sanitation). |
| **Lack of User Interface:** Models exist only in code. | **Streamlit Command Center:** Visual dashboard for health officers. |

## Phase 1: Core Data Cleaning 



```python
import pandas as pd
import numpy as np
from pathlib import Path

DATA_DIR = Path('csv')


# Step 1: Risk Factors (static table)
risk_factors_path = DATA_DIR / 'subcounty_risk_factors.csv'
risk = pd.read_csv(risk_factors_path)

#Standardizing name conventions to ensure exact key matches
for col in ['Sub_County', 'County']:
    if col in risk.columns:
        risk[col] = risk[col].astype(str).str.strip().str.upper()

required_cols = ['Sub_County', 'County', 'WASH_Priority', 'Epi_Priority', 'Risk_Score']
missing = [c for c in required_cols if c not in risk.columns]
if missing:
    raise ValueError(f"Risk factors file missing columns: {missing}. Available: {list(risk.columns)}")

risk_master = risk[required_cols].copy()

risk_master.head()
```

```python

# Step 2: NASA weather data
from io import StringIO

def load_nasa_weather(file_path: Path, county_tag: str) -> pd.DataFrame:
    with open(file_path, 'r', encoding='utf-8', errors='ignore') as f:
        lines = f.readlines()

    marker = '-END HEADER-'
    start_idx = None
    for i, line in enumerate(lines):
        if marker in line:
            start_idx = i + 1
            break

    if start_idx is None:
        raise ValueError(f"Header marker '{marker}' not found in {file_path}")

    csv_text = ''.join(lines[start_idx:])
    df = pd.read_csv(StringIO(csv_text))

    df.columns = [c.strip() for c in df.columns]

    # File tagging
    df['County'] = county_tag

    # Date engineering
    if 'YEAR' not in df.columns or 'DOY' not in df.columns:
        raise ValueError(f"Expected YEAR and DOY columns in {file_path}. Found: {list(df.columns)}")

    df['YEAR'] = pd.to_numeric(df['YEAR'], errors='coerce')
    df['DOY'] = pd.to_numeric(df['DOY'], errors='coerce')

    base = pd.to_datetime(df['YEAR'].astype('Int64').astype(str) + '-01-01', errors='coerce')
    df['Date'] = base + pd.to_timedelta(df['DOY'] - 1, unit='D')

    # Missing value handling
    df = df.replace(-999, np.nan)

    # Weather columns
    id_cols = {'YEAR', 'DOY', 'Date', 'County'}
    weather_cols = [c for c in df.columns if c not in id_cols]

    # Sort + linear interpolation for continuity
    df = df.sort_values('Date')
    for c in weather_cols:
        df[c] = pd.to_numeric(df[c], errors='coerce')

    df[weather_cols] = df[weather_cols].interpolate(method='linear', limit_direction='both')

    keep_cols = ['County', 'Date'] + weather_cols
    return df[keep_cols].copy()

nasa_files = [
    (DATA_DIR / 'Kisumu_raw.csv', 'KISUMU'),
    (DATA_DIR / 'Migori_raw.csv', 'MIGORI'),
    (DATA_DIR / 'Nairobi_raw.csv', 'NAIROBI'),
]

weather_list = [load_nasa_weather(fp, tag) for fp, tag in nasa_files]
weather = pd.concat(weather_list, ignore_index=True)

weather[['County', 'Date']].head(), weather.shape
```

```python

# Step 3: Cholera cases
cases_path = DATA_DIR / 'actual_cases_2026_test.csv'
cases = pd.read_csv(cases_path)

# Standardizing Date
if 'Date' not in cases.columns:
    for alt in ['date', 'DATE', 'Week', 'epi_week', 'Epi_Week']:
        if alt in cases.columns:
            cases = cases.rename(columns={alt: 'Date'})
            break

if 'Date' not in cases.columns:
    raise ValueError(f"Cases file missing 'Date' column. Available: {list(cases.columns)}")

cases['Date'] = pd.to_datetime(cases['Date'], errors='coerce')

# Standardizing Sub_County
if 'Sub_County' not in cases.columns:
    for alt in ['subcounty', 'Subcounty', 'Sub-county', 'SubCounty', 'sub_county', 'SUB_COUNTY']:
        if alt in cases.columns:
            cases = cases.rename(columns={alt: 'Sub_County'})
            break

if 'Sub_County' not in cases.columns:
    raise ValueError(f"Cases file missing 'Sub_County' column. Available: {list(cases.columns)}")

cases['Sub_County'] = cases['Sub_County'].astype(str).str.strip().str.upper()

# Target columns
confirmed_col = None
for c in ['Confirmed_Cases', 'confirmed_cases', 'Confirmed', 'ConfirmedCases', 'Cases', 'case_count']:
    if c in cases.columns:
        confirmed_col = c
        break

if confirmed_col is None:
    raise ValueError(f"Could not find confirmed cases column in {list(cases.columns)}")

cases[confirmed_col] = pd.to_numeric(cases[confirmed_col], errors='coerce').fillna(0)

# Binary Outbreak
cases['Outbreak'] = (cases[confirmed_col] > 0).astype(int)

# Normalizing Confirmed_Cases column name
cases['Confirmed_Cases'] = cases[confirmed_col]

# Aggregating duplicates per Sub_County & Date
cases_clean = (
    cases.groupby(['Sub_County', 'Date'], as_index=False)
         .agg({'Confirmed_Cases': 'sum', 'Outbreak': 'max'})
)

cases_clean.head(), cases_clean.shape
```

```python
risk_master = risk_master
weather = weather
cases_clean = cases_clean
print('risk_master', risk_master.shape, 'weather', weather.shape, 'cases_clean', cases_clean.shape)
```

## Phase 2: Data Merging & 14-Day Lag Engineering



```python

# Step 4: Create the Full Geographic Time-Series Base

# Joining cleaned weather data with risk factors using County
base = weather.merge(risk_master, on='County', how='inner')

# Mandatory sorting
base = base.sort_values(['Sub_County', 'Date'], ascending=[True, True]).reset_index(drop=True)

base.head(), base.shape
```

```python
# Step 5: Engineer Temporal Features (Lags & Rolling Windows(14 days))

# Infer weather column names
def pick_col(df, keywords):
    for k in keywords:
        for c in df.columns:
            if k.lower() in c.lower():
                return c
    return None

rain_col = pick_col(base, ['rain', 'precip', 'precipitation', 'PRECTOT', 'PRECTOTCORR', 'IMERG', 'RFE'])
temp_col = pick_col(base, ['temp', 't2m', 'T2M'])
hum_col  = pick_col(base, ['humidity', 'rh2m', 'RH2M'])

if rain_col is None or temp_col is None or hum_col is None:
    raise ValueError(f"Could not infer weather columns. rain_col={rain_col}, temp_col={temp_col}, hum_col={hum_col}")

lag_days = 14

# Point-in-Time lags (exactly 14 days)
base['rainfall_lag_14'] = base.groupby('Sub_County')[rain_col].shift(lag_days)
base['temp_lag_14'] = base.groupby('Sub_County')[temp_col].shift(lag_days)
base['humidity_lag_14'] = base.groupby('Sub_County')[hum_col].shift(lag_days)

# Rolling windows over the past 14 days (cumulative/averages)
gb = base.groupby('Sub_County', group_keys=False)
base['rainfall_14d_sum'] = gb[rain_col].rolling(window=14, min_periods=14).sum().reset_index(level=0, drop=True)
base['temp_14d_avg'] = gb[temp_col].rolling(window=14, min_periods=14).mean().reset_index(level=0, drop=True)
base['humidity_14d_avg'] = gb[hum_col].rolling(window=14, min_periods=14).mean().reset_index(level=0, drop=True)

print({'rain_col': rain_col, 'temp_col': temp_col, 'hum_col': hum_col})
base[['Sub_County','Date','rainfall_lag_14','rainfall_14d_sum','temp_lag_14','temp_14d_avg','humidity_lag_14','humidity_14d_avg']].head(20)
```

```python
# Step 6: Attach the Target Variable

full = base.merge(
    cases_clean[['Sub_County','Date','Outbreak','Confirmed_Cases']],
    on=['Sub_County','Date'],
    how='left'
)

# Filling nulls with 0
for c in ['Outbreak','Confirmed_Cases']:
    if c in full.columns:
        full[c] = full[c].fillna(0)

full[['Sub_County','Date','Outbreak','Confirmed_Cases']].head(), full.shape
```

```python
# Step 7: Final Cleanup & Pruning

full = full.sort_values(['Sub_County','Date']).reset_index(drop=True)

# Dropping first 14 rows per sub-county group
full = full.groupby('Sub_County').apply(lambda g: g.iloc[14:]).reset_index(drop=True)

# Dropping YEAR and DOY if present
for c in ['YEAR','DOY']:
    if c in full.columns:
        full = full.drop(columns=[c])

print('final shape:', full.shape)
full.head()
```

```python
# Exporting the final dataset for tableau/modeling
output_path = DATA_DIR / 'full_geotemporal_dataset.csv'
full.to_csv(output_path, index=False)
print(f"Exported full dataset to {output_path}")
```
